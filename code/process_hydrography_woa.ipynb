```python
import xarray as xr
import gsw_xarray as gsw

datadir = "../data/"
```

```python
ds = xr.open_mfdataset(datadir + "/originaldata/" + "woa18_decav81B0*.nc",
                     decode_times=False
                    )
ds
```

```python
ds = xr.merge([ds.t_an, ds.s_an, ds.t_dd, ds.s_dd])
ds
```

```python
ds = ds.squeeze().drop_vars("time")
ds
```

```python
ds = ds.sel(lat=slice(60, 90))
ds
```

```python
p = gsw.p_from_z(z=-ds.depth, lat=ds.lat) 
ct = gsw.CT_from_t(SA=ds.s_an, t=ds.t_an, p=p)

ds["t_an"] = ct
ds = ds.rename({"t_an" : "ct", "s_an" : "sa"})
density = gsw.rho(SA=ds.sa, CT=ds.ct, p=p)
density
```

```python
ds.to_netcdf(datadir + "hydrography_woa.nc")
density.to_netcdf(datadir + "density_woa.nc")
```
